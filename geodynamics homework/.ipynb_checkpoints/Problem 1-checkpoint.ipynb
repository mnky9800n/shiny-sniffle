```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
matplotlib.style.use('default')
```

```python
wave_number = np.linspace(0, 7, 1000)
Ra = np.linspace(0,5001, 1000)
growth_rate = (Ra*wave_number**2 - (np.pi**2 + wave_number**2)**3)/(np.pi**2 +wave_number**2)**2
Ra_critical = (np.pi**2 + wave_number**2)**3/(wave_number**2)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(wave_number, growth_rate)
ax.set_xlabel(r'wave number $\frac{2\pi b}{\lambda}$')
ax.set_ylabel(r"growth rate $\frac{\alpha'b^2}{\kappa}$")
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(wave_number, Ra_critical, color='black')
ax.set_yscale('log')
ax.set_xlabel(r'wave number $\frac{2\pi b}{\lambda}$')
ax.set_ylabel(r"$Ra_{cr}$")

ax2 = ax.twinx()

ax2.plot(wave_number, growth_rate, color='red')
ax2.set_ylabel(r"growth rate $\frac{\alpha'b^2}{\kappa}$", color='red')

ax2.plot([0, 7], [0, 0], color='black', linestyle='--')

ra_wvn = np.array([x for x in zip(wave_number, Ra_critical, growth_rate)])
df = pd.DataFrame(ra_wvn, columns=['wave number', 'Ra', 'growth rate'])

wn, ra, a = df[df['growth rate'].between(-0.02, 0.02)].values[0]
ax2.plot([wn, wn], [-10, 40], color='black', linestyle='--')
ax2.text(s='wave number = {wn}\n $Ra_cr$ = {ra}'.format(wn=round(wn,4), ra=round(ra, 2)), x=wn+0.15, y=30)
```
