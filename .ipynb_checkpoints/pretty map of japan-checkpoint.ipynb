```python
% matplotlib inline
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
```

```python
noto_hanto = {'coordinates':(136.686000, 37.220667), 'time':'2007-03-25 00:41:57.909996'
              , 'tz':'UTC', 'magnitude':6.9, 'name':'Noto Hanto'}
western_tottori = {'coordinates':(133.349, 35.274167), 'time':'2000-10-06 04:30:17.940011'
                   , 'tz':'UTC', 'magnitude':6.7, 'name':'Western Tottori'}
sugura_bay = {'coordinates':(138.499333, 34.786167), 'time':'2009-08-10 20:07:05.740017'
              , 'tz':'UTC', 'magnitude':6.5, 'name':'Sugura Bay'}
iwaki = {'coordinates':(140.672667, 36.945667), 'time':'2011-04-11 08:16:12.020008'
         , 'tz':'UTC', 'magnitude':6.6, 'name':'Iwaki'}
iwate_miyage = {'coordinates':(140.880667, 39.029833), 'time':'2008-06-13 23:43:45.359992'
                , 'tz':'UTC', 'magnitude':6.8, 'name':'Iwate Miyage'}
kumamoto = {'coordinates':(130.763, 32.7545), 'time':'2016-04-14 12:26:00'
            , 'tz':'UTC', 'magnitude':7.0, 'name':'Kumamoto'}
tohoku = {'coordinates':(142.861000, 38.103500), 'time':'2011-03-11 05:46:18.119983'
          , 'tz':'UTC', 'magnitude':9.0, 'name':'Tohoku-oki'}
```

```python
def plot_earthquake_on_basemap(eq, m, ax, **kwargs):
    x, y = m(eq['coordinates'][0], eq['coordinates'][1])
    ax.scatter(x, y, s=0.75*np.exp(eq['magnitude']), **kwargs)
    ax.annotate('{year} {name} M{mag}'.format(year=np.datetime64(eq['time']).astype('datetime64[Y]').astype(str)
                                              , mag=str(eq['magnitude']), name=eq['name'])
                , xy=(x, y), xytext=(x, y+0.5), fontsize=15, color='red', fontweight='bold')
```

```python
fig, ax = plt.subplots(1, figsize=(20,20))

m = Basemap(llcrnrlon=127
            ,llcrnrlat=20
            ,urcrnrlon=147
            ,urcrnrlat=47
            , epsg=4269
            , ax=ax
            , resolution='i'
            , area_thresh=1)

m.drawcoastlines()
# m.fillcontinents(color='0.9')

#http://server.arcgisonline.com/arcgis/rest/services
#EPSG Number of America is 4269

# m.arcgisimage(service='ESRI_Imagery_World_2D', xpixels = 9000, verbose= True)

# labels = np.arange(0, 180, 5)
# m.drawparallels(labels, labels)

# fujisan = (138.7278, 35.3606)
# x, y = m(fujisan[0], fujisan[1])
# x2, y2 =m(fujisan[0]+.1, fujisan[1])
# ax.annotate('Mount Fuji', xy=(x,y), xytext=(x2, y2), xycoords='data', color='White'
#             , textcoords='data', arrowprops=dict(arrowstyle="->", color='white'))

plot_earthquake_on_basemap(eq=noto_hanto, m=m, ax=ax, alpha=0.5, color='None', edgecolor='red'
                          ,linewidth=2, zorder=5)


plot_earthquake_on_basemap(eq=western_tottori, m=m, ax=ax, alpha=0.5, color='None', edgecolor='red'
                          ,linewidth=2, zorder=5)


plot_earthquake_on_basemap(eq=sugura_bay, m=m, ax=ax, alpha=0.5, color='None', edgecolor='red'
                          ,linewidth=2, zorder=5)


plot_earthquake_on_basemap(eq=iwate_miyage, m=m, ax=ax, alpha=0.5, color='None', edgecolor='red'
                          ,linewidth=2, zorder=5)


plot_earthquake_on_basemap(eq=kumamoto, m=m, ax=ax, alpha=0.5, color='None', edgecolor='red'
                          ,linewidth=2, zorder=5)

plot_earthquake_on_basemap(eq=iwaki, m=m, ax=ax, alpha=0.5, color='None', edgecolor='red'
                          ,linewidth=2, zorder=5)


# ax2 = fig.add_axes([0.15, 0.25, 0.1, 0.2])
# x = np.arange(0,2*np.pi, 0.01)
# y = np.sin(x)
# ax2.plot(x, y)
# ax2.set_aspect(3)

ax.axis('off')

fig.savefig('japan.png', dpi=500, bbox_inches='tight', transparent=True)
```
