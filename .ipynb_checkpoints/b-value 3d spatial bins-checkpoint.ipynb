```python
%matplotlib inline
import pandas as pd
import numpy as np
from utilities import plotting
from utilities import stats
from utilities import timestamps
from utilities import import_export
from utilities import polygon_selection
from mpl_toolkits.basemap import Basemap
from itertools import product
import matplotlib.pyplot as plt
```

```python
df = import_export.import_catalog(location='data/ccu.dat', delimiter='\t', names=['lon', 'lat', 'decimal_year', 'month', 'day', 'mag'
                                                        , 'depth', 'hour', 'minute', 'second', 'horizontal_error'
                                                        , 'depth_error', 'mag_err'])
kumamoto = {'coordinates':(130.763, 32.7545), 'time':'2016-04-14 12:26:00', 'tz':'UTC', 'magnitude':7.0}
eq = kumamoto

df_regional = polygon_selection.get_node_data(data=df, node=eq['coordinates'], radius=75, m=1)
df_regional = df_regional.loc[df_regional.index < eq['time']]
df_regional = df_regional[df_regional.depth <=25]
df_regional = df_regional[df_regional.mag >= 0.5]
```

```python
plotting.plot_seismicity_map(df_regional)
```

```python
fig, ax = plt.subplots()
plotting.plot_seismicity_rate(df_regional[df_regional.decimal_year.between(1999, 2002)], fig, ax)
```

```python
radii = np.arange(5, 26, 1)
timii = pd.date_range(start='01-01-1990 00:00:00', end=eq['time'], freq='12M', tz='GMT')
depths = np.arange(0, 25, 4)
parameters = [x for x in product(radii, timii, depths)]

increment = 0.05
lat_min, lat_max = (32, 33.5)
lats = np.linspace(32, 33.5, (33.5-32)/increment + 1)
lons = np.linspace(129.5, 132, (132 - 129.5)/increment + 1)
lats_lons = [x for x in product(lons, lats)]

len(lats_lons)
everything = []

for lon_lat in lats_lons:
    everything.append((df_regional, parameters, lon_lat))
```

```python
def calc_b_value_for_parameters(dataframe, parameters, location):
    rows = []
    for r, t, d in parameters:
        try:
            df_c = polygon_selection.get_node_data(node=location, data=dataframe[dataframe.depth.between(d-4, d+4)], radius=r, m=1)
            df_c = df_c.loc[df_c.index >= t]

            a, b, bstd, n, mc = stats.calc_fmd_stats_with_mc(df_c.mag)
            row = np.array((r,) + (t,) + (a, b, bstd, n, mc) + tuple(location))
            rows.append(row)
        except:
            row = np.array((r,) + (t,) +(np.nan, np.nan, np.nan, np.nan, np.nan) + tuple(location))
    
    return rows

def calc_b_for_parameters_attr_iter(attr):
    dataframe, parameters, location = attr
    return calc_b_value_for_parameters(dataframe, parameters, location)
```

```python
%timeit calc_b_value_for_parameters(df_regional, [[5, timii[0], 5]], eq['coordinates'])
```

```python
from multiprocessing import Pool
pool = Pool()

results = pool.map(calc_b_for_parameters_attr_iter, everything)
```

```python
results[0]
```

```python
results_df = [pd.DataFrame(r, columns=['radius', 'start_time', 'a', 'b', 'bstd', 'n', 'mc', 'lon', 'lat']) for r in results]
```

```python
depth_centers = np.array(everything[0][1])[:,2]
for r in results_df:
    r['depth center'] = depth_centers.astype(int)
```

```python
results_df = pd.concat(results_df)
```

```python
results_avg = results_df.groupby(['depth center', 'start_time', 'lon', 'lat']).mean().reset_index()
```

```python
results_masked = results_avg.copy()
results_masked.ix[results_masked.bstd >= 0.05, 'b'] = np.nan
results_masked.ix[results_masked.bstd == 0.0, 'b'] = np.nan
results_masked.ix[results_masked.n < 50, 'b'] = np.nan
results_masked.b.hist(bins=100, range=(0.5, 1.5), log=True)
```

```python
st = '01-31-2001'

for d in results_avg['depth center'].unique():
    df_depth = results_masked[(results_masked['start_time'] == st) & (results_masked['depth center'] == d)]

    fig, ax = plt.subplots(figsize=(8, 8))


    lons = results_masked.lon.values
    lats = results_masked.lat.values

    lat_min = lats.min()
    lat_max = lats.max()
    lon_min = lons.min()
    lon_max = lons.max()

    lon_i = np.linspace(lon_min, lon_max, (lon_max - lon_min)/0.05 + 1)
    lat_i = np.linspace(lat_min, lat_max, (lat_max - lat_min)/0.05 + 1)


    m = Basemap(projection='merc'
               ,llcrnrlat=lat_min
               ,urcrnrlat=lat_max
               ,llcrnrlon=lon_min
               ,urcrnrlon=lon_max
               ,resolution='i'
               ,area_thresh=1000
               ,ax=ax)

    m.drawcoastlines()

    data = df_depth.copy()

    n_remove = 250
    # print('shape', data.shape)
    # print('data copy', data[data.n_x < n_remove].b_x.describe())
    data.ix[data.n < n_remove, 'b'] = np.nan
#     data.ix[data.bstd >= 0.05] = np.nan
    # data.ix[data.n_y < n_remove, 'b_y'] = np.nan
    # print('data removal', data[data.n_x < n_remove].b_x.describe())

    zi = data.pivot(index='lat', columns='lon', values='b').values

    xi, _yi = m(lon_i, np.zeros_like(lon_i))
    _xi, yi = m(np.zeros_like(lat_i), lat_i)

    x, y = m(lons, lats)

    xi, yi = np.meshgrid(xi, yi)

    zi = np.ma.masked_invalid(zi)
    # >>> ma.masked_where(a <= 2, a)
    # zi = np.ma.masked_where(zi > 2.0, zi)
    # zi = np.ma.masked_where(zi < 0.5, zi)

    cbar = ax.pcolormesh(xi, yi, zi, cmap='jet', vmin=0.5, vmax=1.2, alpha=0.95)

    fig.colorbar(cbar, label='b', orientation='horizontal')
    ax.set_title('d={d}km '.format(d=d) + str(data.start_time.head(1)[0]))

    x, y = m(eq['coordinates'][0], eq['coordinates'][1])
    ax.scatter(x, y, s=500, color='None', edgecolor='black')
    
    m5lon, m5lat = df_regional[(df_regional.mag>4.9) & (df_regional.decimal_year.between(1999, 2001))][['lon','lat']].values[0]
    x, y = m(m5lon, m5lat)
    ax.scatter(x, y, s=250, color='None', edgecolor='black')
    
    plt.tight_layout()
    
    fig.savefig('3d_maps/kumamoto_{d}_{st}_masked.png'.format(d=d, st=st), bbox_inches='tight')
    plt.close()
```

```python
results_2000 = results_masked[results_masked['start_time'] == '01-31-2000']
results_2001 = results_masked[results_masked['start_time'] == '01-31-2001']
results_2010 = results_masked[results_masked['start_time'] == '01-31-2010']
```

```python
results_diff = results_2000.merge(results_2010, on=['depth center', 'lon', 'lat']).merge(results_2001, on=['depth center', 'lon', 'lat'])
```

```python
results_diff['bdiff_2000_2010'] = -1.*(results_diff.b_x - results_diff.b_y)
results_diff['bdiff_2000_2001'] = -1.*(results_diff.b_x - results_diff.b)
results_diff['bdiff_2001_2010'] = -1.*(results_diff.b - results_diff.b_y)
```

```python
st = '01-31-2001'
d = 4
for d in results_avg['depth center'].unique():
    df_depth = results_diff[(results_diff['depth center'] == d)]

    fig, ax = plt.subplots(figsize=(8, 8))


    lons = results_avg.lon.values
    lats = results_avg.lat.values

    lat_min = lats.min()
    lat_max = lats.max()
    lon_min = lons.min()
    lon_max = lons.max()

    lon_i = np.linspace(lon_min, lon_max, (lon_max - lon_min)/0.05 + 1)
    lat_i = np.linspace(lat_min, lat_max, (lat_max - lat_min)/0.05 + 1)


    m = Basemap(projection='merc'
               ,llcrnrlat=lat_min
               ,urcrnrlat=lat_max
               ,llcrnrlon=lon_min
               ,urcrnrlon=lon_max
               ,resolution='i'
               ,area_thresh=1000
               ,ax=ax)

    m.drawcoastlines()
    m.fillcontinents('0.92', zorder=0)
    data = df_depth.copy()

    n_remove = 250
    # print('shape', data.shape)
    # print('data copy', data[data.n_x < n_remove].b_x.describe())
    data.ix[data.n_x < n_remove, 'bdiff_2001_2010'] = np.nan
    data.ix[data.n_y < n_remove, 'bdiff_2001_2010'] = np.nan
    data.ix[data.b]
    # print('data removal', data[data.n_x < n_remove].b_x.describe())

    zi = data.pivot(index='lat', columns='lon', values='bdiff_2001_2010').values

    xi, _yi = m(lon_i, np.zeros_like(lon_i))
    _xi, yi = m(np.zeros_like(lat_i), lat_i)

    x, y = m(lons, lats)

    xi, yi = np.meshgrid(xi, yi)

    zi = np.ma.masked_invalid(zi)
    # >>> ma.masked_where(a <= 2, a)
    # zi = np.ma.masked_where(zi > 2.0, zi)
    # zi = np.ma.masked_where(zi < 0.5, zi)

    cbar = ax.pcolormesh(xi, yi, zi, cmap='seismic', vmin=-0.15, vmax=0.15, alpha=0.95)

    fig.colorbar(cbar, label='b', orientation='horizontal')
    ax.set_title('d={d}km '.format(d=d) + str(data.start_time_x.head(1)[0]) + '   ' + str(data.start_time_y.head(1)[0]))

    x, y = m(eq['coordinates'][0], eq['coordinates'][1])
    ax.scatter(x, y, s=500, color='None', edgecolor='black')
    
    m5lon, m5lat = df_regional[(df_regional.mag>4.9) & (df_regional.decimal_year.between(1999, 2001))][['lon','lat']].values[0]
    x, y = m(m5lon, m5lat)
    ax.scatter(x, y, s=250, color='None', edgecolor='black')

    fig.savefig('3d_maps/kumamoto_bdiff_{d}_{st}_bdiff_2001_2010_masked.png'.format(d=d, st=st), bbox_inches='tight')
    plt.close()
```
